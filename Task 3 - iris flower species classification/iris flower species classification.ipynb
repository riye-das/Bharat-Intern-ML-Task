# data pre processing

```python
iris.drop("Id",axis=1,inplace =True) 
```

```python
x=iris.drop(['Species'],axis=1)
y =iris['Species']
print(x)
print(y)
```

# correlation matrix

```python
corr = iris.corr()
print(corr)
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(corr, annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
y = le.fit_transform(y)
y
```

```python
x = np.array(x)
x
```

```python
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print(x_train)
print(x_train.size)
```

```python
print(x_test)
print(x_test.size)
```

```python
### LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', multi_class='auto')
model.fit(x_train, y_train)
prediction = model.predict(x_test)
print(accuracy_score(y_test, prediction)*100)
```

```python
plt.scatter(y_test,prediction,c='yellow', edgecolor='black')
plt.show()
```

```python
##  Decission Tree
```

```python
from sklearn import tree
DTree = tree.DecisionTreeClassifier()
DTree.fit(x_train,y_train)
prediction_DTree = DTree.predict(x_test)
accuracy_DTree = accuracy_score(y_test,prediction_DTree)*100
print(accuracy_DTree)
```

```python
plt.scatter(y_test,prediction_DTree,c='yellow', edgecolor='black')
plt.show()
```
